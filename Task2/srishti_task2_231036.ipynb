```python
dataset_dir = '/kaggle/input/ravdess-8k'


speech_dir = os.path.join(dataset_dir, 'ravdess_rewritten_8k')
#noise_dir = os.path.join(dataset_dir, 'noise_files')
#output_dir = '/kaggle/working/combined_files'
data_dir = '/kaggle/input/urban-sound-8k'
noise_dir = os.path.join(data_dir, 'urbansound_8k')

```

```python
speech_files = [f for f in os.listdir(speech_dir) if os.path.isfile(os.path.join(speech_dir, f))]
noise_files = [f for f in os.listdir(noise_dir) if os.path.isfile(os.path.join(noise_dir, f))]
```

```python
def combine_audio_files(speech_file_path, noise_file_path, output_file_path):
    speech, sr_speech = librosa.load(speech_file_path, sr=None)
    noise, sr_noise = librosa.load(noise_file_path, sr=None)
    combined_audio = speech + noise
    combined_audio = combined_audio / np.max(np.abs(combined_audio))
    sf.write('/kaggle/working/', combined_audio, sr_speech)
    print(f"Combined audio saved to output_path")
```

```python
speech_files = sorted(os.listdir(speech_dir))
noise_files = sorted(os.listdir(noise_dir))
for speech_file, noise_file in zip(speech_files, noise_files):
    speech_file_path = os.path.join(speech_dir, speech_file)
    noise_file_path = os.path.join(noise_dir, noise_file)
    output_file_path = os.path.join(output_dir, f'combined_{speech_file}')
```

```python
for speech_file in speech_files:
    for noise_file in noise_files:
        speech_path = os.path.join(speech_dir, speech_file)
        noise_path = os.path.join(noise_dir, noise_file)
        
        # Load the speech and noise files
        speech, sr_speech = librosa.load(speech_path, sr=None)
        noise, sr_noise = librosa.load(noise_path, sr=None)
        
        # Ensure both audio files have the same sampling rate
        if sr_speech != sr_noise:
            print(f"Sampling rates differ: {sr_speech} (speech) vs {sr_noise} (noise)")
            continue  # Skip this pair if sampling rates don't match

        # Truncate or pad noise to match the length of the speech file
        if len(noise) < len(speech):
            noise = librosa.util.fix_length(noise, len(speech))
        else:
            noise = noise[:len(speech)]

        # Combine speech and noise
        combined = speech + noise
        
        # Define the output file path
        output_file = f"combined_{os.path.splitext(speech_file)[0]}_{os.path.splitext(noise_file)[0]}.wav"
        output_path = os.path.join(output_dir, output_file)
        
        # Save the combined audio file
        sf.write(output_path, combined, sr_speech)
        
        print(f"Saved combined file: {output_path}")
```
